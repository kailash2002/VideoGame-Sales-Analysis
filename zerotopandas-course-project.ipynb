## Exploratory Analysis and Visualization

Let's begin by importing`matplotlib.pyplot` and `seaborn`.

````python
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

sns.set_style('darkgrid')
matplotlib.rcParams['font.size'] = 14
matplotlib.rcParams['figure.figsize'] = (9, 5)
matplotlib.rcParams['figure.facecolor'] = '#00000000'
````

````python
G=sales_df1.Genre.unique()
G.sort()
G
````

````python
G_sales=sales_df.groupby('Genre')['Global_Sales'].sum()
G_sales
````

````python
plt.figure(figsize=(16,8))
plt.xticks(rotation=120)
plt.xlabel('Genre')
plt.ylabel('sales_in_millions')
plt.title('most usable game')
sns.barplot(G,G_sales);
````

as  we can see the above graph sports  mostly use type of genre

````python
na_sales=sales_df1.NA_Sales.sum()
pal_sales=sales_df1.PAL_Sales.sum()
other_sales=sales_df1.Other_Sales.sum()
jp_sales=sales_df1.JP_Sales.sum()

s=[na_sales,pal_sales,other_sales,jp_sales]
````

````python
labels='NA_Sales','PAL_Sales','JP_Sales','Other_Sales'
plt.title('percentage of sale')
plt.pie(s,labels=labels,autopct='%.2f%%',shadow=True,explode=[0,0.1,0.03,0.09]);
````

````python
na_sales5=sales_latest_df.NA_Sales.sum()
pal_sales5=sales_latest_df.PAL_Sales.sum()
other_sales5=sales_latest_df.Other_Sales.sum()
jp_sales5=sales_latest_df.JP_Sales.sum()

s5=[na_sales5,pal_sales5,other_sales5,jp_sales5]
````

````python
labels='NA_Sales_latest','PAL_Sales_latest','JP_Sales_latest','Other_Sales_latest'
plt.title('percentage of sale')
plt.pie(s5,labels=labels,autopct='%.2f%%',shadow=True,explode=[0,0.1,0.03,0.09]);
````

#### Q2: which are the top 5 publicers of games in respect to number of game produced?

````python
p=sales_df1['Publisher'].value_counts().head(5)
````

````python
p
````

````python
plt.figure(figsize=(16,8))
plt.xticks(rotation=120)
sns.barplot(p.index,p)
plt.title('maximum number of game produced')
````

above graoh are clearly shown which publisher are most publish game

#### Q3: which year had maximum global sales?

````python
game_s=sales_df1.groupby('Year')['Global_Sales'].sum()
````

````python
game_s
````

````python
plt.figure(figsize=(16,8))
plt.xticks(rotation=120)
plt.xlabel('years')
plt.ylabel('total sales')
sns.barplot(game_s.index,game_s)
plt.title('maximum global sales');
````

as we can see from the above bar graph maximun sales of video game was in years 2009,2010,2011,2008

#### Q4: which is most preffered platform by companises to makes game for in last five years?

````python
k=sales_latest_df.Platform.value_counts()
print(k)
````

````python
plt.figure(figsize=(16,8))
plt.xticks(rotation=120)
plt.xlabel('platforms')
plt.ylabel('no of games')
sns.barplot(k.index,k)
plt.title('most preferred platform to make game in last five years');
````

as we can see from the above bar graph the most preferred platforms to make games . in last five years is pc followed by 3DS,PS3,PS4 and so on.

#### Q5: which game have most sales worldwide?
